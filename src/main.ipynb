## Búsqueda Lineal (Linear Search)

La búsqueda lineal es el algoritmo de búsqueda más simple. Recorre cada elemento del array uno por uno hasta encontrar el elemento buscado. Si el elemento no está en el array, el algoritmo recorre todo el array sin encontrarlo.

### Algoritmo
1. Comienza desde el primer elemento del array.

2. Compara el elemento actual con el valor buscado.

3. Si son iguales, devuelve la posición del elemento.

4. Si no son iguales, avanza al siguiente elemento.

5. Repite los pasos 2-4 hasta encontrar el elemento o llegar al final del array.

### Complejidad Algorítmica
- Peor caso: $O(n)$

    El peor caso ocurre cuando el elemento buscado está al final del array o no está presente. En este caso, el algoritmo recorre todos los nn elementos.

- Mejor caso: $O(1)$

    El mejor caso ocurre cuando el elemento buscado está en la primera posición del array.

- Caso promedio: $O(n)$

    En promedio, el algoritmo recorre la mitad del array ($\frac{n}{2}$).

### Características
- Recorre cada elemento del array uno por uno.
- Funciona en arrays ordenados y desordenados.
- Simple de implementar, pero ineficiente para arrays grandes.

## Búsqueda Binaria (Binary Search)

La búsqueda binaria es un algoritmo eficiente para buscar un elemento en un array ordenado. Divide repetidamente el array en dos partes y descarta la mitad en la que el elemento buscado no puede estar.

### Algoritmo
1. Define dos punteros: `low` (inicio del array) y `high` (final del array).

2. Calcula el punto medio (`mid`) entre `low` y `high`.

3. Compara el elemento en mid con el valor buscado:

    - Si son iguales, devuelve la posición mid.

    - Si el valor buscado es menor, descarta la mitad derecha y ajusta `high = mid - 1`.

    - Si el valor buscado es mayor, descarta la mitad izquierda y ajusta `low = mid + 1`.

4. Repite los pasos 2-3 hasta encontrar el elemento o hasta que `low` sea mayor que `high`.

### Complejidad Algorítmica
- Peor caso: $O(log (n))$

    El peor caso ocurre cuando el elemento no está en el array o está en una posición que requiere dividir el array hasta que solo quede un elemento.

- Mejor caso: $O(1)$

    El mejor caso ocurre cuando el elemento buscado está justo en el medio del array.

- Caso promedio: $O(log (n))$

    En promedio, el algoritmo divide el array en dos partes en cada iteración.

### Características
- Requiere que el array esté ordenado.
- Divide el array en dos partes en cada iteración.
- Muy eficiente para arrays grandes.

## Búsqueda Ternaria (Ternary Search)

La búsqueda ternaria es una variante de la búsqueda binaria que divide el array en tres partes en lugar de dos. Es útil para buscar en arrays ordenados, pero en la práctica es menos eficiente que la búsqueda binaria debido a más comparaciones por iteración.

### Algoritmo
1. Define dos punteros: `left` (inicio del array) y `right` (final del array).

2. Divide el array en tres partes:

    - `mid1 = left + (right - left) // 3`

    - `mid2 = right - (right - left) // 3`

3. Compara el valor buscado con los elementos en `mid1` y `mid2`:

    - Si el valor buscado es igual a `arr[mid1]`, devuelve `mid1`.

    - Si el valor buscado es igual a `arr[mid2]`, devuelve `mid2`.

    - Si el valor buscado es menor que `arr[mid1]`, descarta la parte derecha y ajusta `right = mid1 - 1`.

    - Si el valor buscado es mayor que `arr[mid2]`, descarta la parte izquierda y ajusta `left = mid2 + 1`.

    - Si el valor buscado está entre `arr[mid1]` y `arr[mid2]`, ajusta `left = mid1 + 1` y `right = mid2 - 1`.

4. Repite los pasos 2-3 hasta encontrar el elemento o hasta que `left` sea mayor que `right`.

### Complejidad Algorítmica
- Peor caso: $O(log_⁡3(n))$

    El peor caso ocurre cuando el elemento no está en el array o está en una posición que requiere dividir el array hasta que solo quede un elemento.

- Mejor caso: $O(1)$

    El mejor caso ocurre cuando el elemento buscado está justo en mid1 o mid2.

- Caso promedio: $O(log_⁡3(n))$

    En promedio, el algoritmo divide el array en tres partes en cada iteración.

### Características
- Similar a la búsqueda binaria, pero divide el array en tres partes.
- Requiere que el array esté ordenado.
- Ligeramente más lenta que la búsqueda binaria debido a más comparaciones por iteración.
